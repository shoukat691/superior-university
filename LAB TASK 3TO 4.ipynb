```python
# LAB TASK 3
class ModelBasedReflexAgent:
    def _init_(self, desired_temperature):
        self.desired_temperature = desired_temperature
        self.previous_action = {}  # This will store the previous action for each room

    def perceive(self, current_temperature):
        return current_temperature

    def act(self, room, current_temperature):
        # Check previous action for the room
        prev_action = self.previous_action.get(room, None)

        if current_temperature < self.desired_temperature:
            action = "Turn on heater"
        else:
            action = "Turn off heater"

        # If the action is the same as the previous one, avoid redundant action
        if action == prev_action:
            result = f"No action needed. Heater is already {prev_action.split()[-1]}."
        else:
            result = action
            self.previous_action[room] = action  # Update the previous action

        return result

# Simulating different rooms with different current temperatures
rooms = {
    "Living Room": 18,
    "Bedroom": 22,
    "Kitchen": 20,
    "Bathroom": 24
}

# Desired temperature for all rooms
desired_temperature = 22
agent = ModelBasedReflexAgent(desired_temperature)

# Run the agent for each room and simulate consecutive cycles
for _ in range(2):  # Simulate two cycles to observe the behavior of the agent
    print("\nCycle:")
    for room, temperature in rooms.items():
        action = agent.act(room, temperature)
        print(f"{room}: Current temperature = {temperature}Â°C. {action}.")
        
```

```python
#LAB TASK 4 
# QUESTION no 1
def luhn_algorithm(card_number):
    # Reverse the card number string and convert each digit to an integer
    digits = [int(digit) for digit in card_number[::-1]]
    
    checksum = 0
    
    # Loop through the digits
    for i, digit in enumerate(digits):
        if i % 2 == 1:  # Double every second digit
            digit *= 2
            if digit > 9:  # If doubling results in a number greater than 9, subtract 9
                digit -= 9
        checksum += digit

    # Card is valid if checksum modulo 10 is equal to 0
    return checksum % 10 == 0

# Example usage:
card_number = "4532015112830366"
print("Valid Card" if luhn_algorithm(card_number) else "Invalid Card")
```

```python
# QUESTION NO 2
def remove_punctuation(input_string):
    # Define punctuations that should be removed
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    
    result = ""
    for char in input_string:
        if char not in punctuations:
            result += char
    return result

# Example usage:
user_input = "Hello, World! This is a test string."
clean_string = remove_punctuation(user_input)
print(clean_string)
```

```python
# QUESTION NO 3
def sort_words(text):
    words = text.split()  # Split the string into words
    # Implementing a simple bubble sort to sort the words
    n = len(words)
    for i in range(n):
        for j in range(0, n-i-1):
            if words[j].lower() > words[j+1].lower():  # Compare words ignoring case
                words[j], words[j+1] = words[j+1], words[j]
    return ' '.join(words)

# Example usage:
text = "Sorting words in alphabetical order without using sort function"
sorted_text = sort_words(text)
print(sorted_text)
```
